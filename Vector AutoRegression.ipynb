```python
import pandas as pd
# Vector AutoRegression
from statsmodels.tsa.vector_ar.var_model import VAR
from random import random
import numpy as np
from sqlalchemy import create_engine
```

```python
# Example of Vector AutoRegresion
# contrived dataset with dependency
data = list()
for i in range(100):
    v1 = i + random()
    v2 = v1 + random()
    row = [v1, v2]
    data.append(row)
# fit model
model = VAR(data)
model_fit = model.fit()
# make prediction
yhat = model_fit.forecast(model_fit.y, steps=1)
print(yhat)
```
